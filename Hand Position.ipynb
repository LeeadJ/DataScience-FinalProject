### Creating the dataframes:
- We want to creat a df for the 'sync' and 'spontan' dataframes and one for the alone.
- First we will creat the 'sync' and 'spontan', we will merge the left and right hand with the same FrameID.
- Next we will creat the 'alone' by merging its left hand with the 'Hand_Right' df and remove hands that are misplaced.

```python
def df_creatS(List):
    for i in range(len(List)):
        df = List[i]
        time_subtract = math.ceil(df["Time"][0]) + 7
        df = df[df["Time"] > time_subtract]
        hand = df.groupby(df[' Hand Type'])
        right = hand.get_group("right")
        left = hand.get_group("left")
        df = pd.merge(left, right, how='left', on= ' Frame ID')
        List[i] = df
    return List 
```

```python
Spontan_List_Train = df_creatS(Spontan_List_Train)
Sync_List_Train = df_creatS(Sync_List_Train)

Spontan_List_Test = df_creatS(Spontan_List_Test)
Sync_List_Test = df_creatS(Sync_List_Test)
```

```python
def df_creatAlone(List):
    Hand_Right = pd.read_csv(r"C:\Users\Leead\Desktop\Hands\HandRight.csv")
    for i in range(len(List)):
        df = List[i]
        time_subtract = math.ceil(df["Time"][0]) + 7
        df = df[df["Time"] > time_subtract]
        df = df[df[' Hand Type'] == 'left']
        Hand_Right = Hand_Right[Hand_Right[' Hand Type']=='right']   
        Hand_Right = Hand_Right.iloc[:len(df)]
        Hand_Right = Hand_Right.reset_index(drop=True)
        df = df.reset_index(drop=True)
        df = pd.merge(df,Hand_Right,left_index=True,right_index=True,how='left') 
        List[i] = df
        
    return List 
```

```python
Alone_List_Train = df_creatAlone(Alone_List_Train)

Alone_List_Test = df_creatAlone(Alone_List_Test)
```
